## 10. To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?

```python
top_10_run_getters = df.groupby('batsman')['total_runs'].sum().nlargest(10).index

# Filtering the DataFrame for the top 10 run-getters
top_10_batsmen_df = df[df['batsman'].isin(top_10_run_getters)]

# Grouping by batsman and calculate the total runs for each match
total_runs_by_batsman = top_10_batsmen_df.groupby(['batsman', 'match_id'])['total_runs'].sum().reset_index()

# Grouping by batsman and calculate the average runs per match
average_runs_per_match = total_runs_by_batsman.groupby('batsman')['total_runs'].mean().sort_values(ascending=False)
```

```python
average_runs_per_match
```

```python
plt.figure(figsize=(12, 6))
average_runs_per_match.plot(kind='line', marker='o', color='b')
plt.title('Consistent Top 10 Run-Getters')
plt.xlabel('Batsmen')
plt.ylabel('Average Runs per Match')
plt.xticks(rotation = 90)
plt.show()
```

## 11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?

```python
matches_2019 = df[df['season'] == 2019]

# Filtering for runs scored through 6's and 4's
runs_6s_4s = matches_2019[(matches_2019['batsman_runs'] == 6) | (matches_2019['batsman_runs'] == 4)]

# Grouping by batsman and calculate the total runs through 6's and 4's for each batsman
total_runs_6s_4s_by_batsman = runs_6s_4s.groupby('batsman')['total_runs'].sum().sort_values(ascending=False)

# Getting the batsman who scored the highest number of runs through 6's and 4's
top_batsman_6s_4s = total_runs_6s_4s_by_batsman.idxmax()

# Geting the total runs scored by the top batsman
total_runs_by_top_batsman = total_runs_6s_4s_by_batsman.max()
```

## 20. which player has the highest hard-hitting ability?

```python
boundary_runs = df[(df['total_runs'] == 4) | (df['total_runs'] == 6)]

# Grouping by batsman and calculate the count of boundaries and the total runs and balls faced
player_stats = df.groupby('batsman').agg({
    'total_runs': 'sum',
    'ball': 'count',
    'total_runs': 'sum',
    'season': 'nunique'  
})

# Filtering out players who haven't played in multiple seasons (optional, adjust as needed)
player_stats = player_stats[player_stats['season'] > 10]

# Calculating hard-hitting ability as the ratio of boundaries to total balls faced
player_stats['hard_hitting_ability'] = (boundary_runs.groupby('batsman')['total_runs'].count() / player_stats['ball']) * 100

# Getting the player with the highest hard-hitting ability
player_highest_hard_hitting_ability = player_stats['hard_hitting_ability'].idxmax()
highest_hard_hitting_ability = player_stats['hard_hitting_ability'].max()
```
